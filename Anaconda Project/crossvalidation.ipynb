```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df
```

```python
df["species"].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species',axis=1)
```

```python
y=df['species']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=15)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(
    X_test,y_test
)
```

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(kernel='linear',C=10,gamma='auto')
```

```python

model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
#cross validation is reduce overfitting situation
from sklearn.model_selection import GridSearchCV
```

```python
classifier=GridSearchCV((model_svm),{
    'kernel':('linear', 'rbf'),
    'C':[1, 10,20,30]
},cv=5,return_train_score=False)
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.cv_results_
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
classifier2=GridSearchCV((model),{
    'weights':('uniform','distance'),
    'n_neighbors':[1,3,5,7,9],
    'algorithm':('auto', 'ball_tree', 'kd_tree', 'brute')
},cv=5,return_train_score=False)
```

```python
classifier2.fit(X_train,y_train)
```

```python
classifier2.cv_results_
```

```python
results=pd.DataFrame(classifier2.cv_results_)
```

```python
results
```

```python
results[['param_algorithm',	'param_n_neighbors','param_weights','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r=RandomizedSearchCV((model_svm),{
    'kernel':('linear', 'rbf'),
    'C':[1, 10,20,30]
},n_iter=4,cv=5,return_train_score=False)
```

```python
classifier_r.fit(X_train,y_train)
```

```python
results=pd.DataFrame(classifier_r.cv_results_)
```
